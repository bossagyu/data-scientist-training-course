```python
# Scipyの基礎
import scipy.linalg as linalg
from scipy.optimize import minimize_scalar
```

```python
# 2-3-2-1 行列式と逆行列の計算

matrix = np.array([[1,-1, -1], [-1,1,-1], [-1,-1,1]])

print('行列式:', linalg.det(matrix))
```

```python
# 逆行列を計算
inverse_matrix = linalg.inv(matrix)
print ('逆行列:', inverse_matrix)

print(matrix.dot(inverse_matrix))
```

```python
# 2-3-2-2 固有値と固有ベクトル
eig_values, eig_vectors = linalg.eig(matrix)
print('固有値:', eig_values)
print('固有ベクトル:', eig_vectors)
```

```python
# 2-3-3-1 方程式の解を求める
def my_function(x):
    return (x**2 + 2*x + 1)

#  ニュートン法の読み込み
from scipy.optimize import newton
print(newton(my_function,0))
```

```python
print(minimize_scalar(my_function, method='Brent'))
```

```python
# 散布図
random.seed(0)

# x軸のデータ
x = np.random.random(30)

# y軸のデータ
y = np.random.random(30)

# グラフのサイズ指定
plt.figure(figsize=(20, 6))

# グラフの描画
plt.plot(x, y, 'o')

# グラフのタイトル
plt.title('Titlle of the Scatter Plot')

# x軸のラベル
plt.xlabel('x')

# y軸のラベル
plt.ylabel('y')

# グリッド線の表示
plt.grid(True)
```

```python
# 連続曲線
numpy_data_x = np.arange(1000)

# 乱数の発生と積み上げ
numpy_data_y = np.random.randn(1000).cumsum()

plt.figure(figsize=(20, 6) )

plt.plot(numpy_data_x, numpy_data_y, label='Random Walk')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
```
